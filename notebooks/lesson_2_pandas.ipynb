мы тоже можем придумывать признаки

```python
def age_category(age):
    """
    < 30 -> 1
    >= 30, <55 -> 2
    >= 55 -> 3
    """
    if age < 30:
        return 1
    elif age < 55:
        return 2
    else:
        return 3
```

```python
age_categories = [age_category(age) for age in passengers_df['Age']]
```

```python
passengers_df["Age_category"] = age_categories
```

```python
passengers_df["Age_category"] = passengers_df["Age"].apply(age_category)
```

# ДЗ


Подсчитать в одном запросе распределение по посещеням в день (toStartOfDay) для мобильного (IsMobile = 1) и не мобильного (IsMobile = 0) траффика c 2014-03-16 21:00:00 по 2014-03-19 16:00:00


```python
result = %sql select t1.cnt as mob, t2.cnt as des, t1.day \
from (SELECT count(Distinct UserID) as cnt, toStartOfDay(EventTime) as day \
      FROM datasets.hits_v1 as h \
      where EventTime between '2014-03-16 21:00:00' and '2014-03-19 16:00:00' \
        and IsMobile = 0 \
      group by day \
      order by day desc) as t1 \
         join (SELECT count(Distinct UserID) as cnt, toStartOfDay(EventTime) as day \
               FROM datasets.hits_v1 as h \
               where EventTime between '2014-03-16 21:00:00' and '2014-03-19 16:00:00' \
                 and IsMobile = 1 \
               group by day \
               order by day desc) as t2 on t1.day = t2.day \

df = result.DataFrame()
df
```
